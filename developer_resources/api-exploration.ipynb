````python
import logging
import os
from pathlib import Path

from dsd_pythonanywhere.client import PythonAnywhereClient

logging.basicConfig(
    level=logging.DEBUG, force=True, format="%(asctime)s - %(levelname)s - %(message)s"
)
````

````python
# VS Code's Jupyter extension doesn't support loading .envrc, so do it manually here

envrc = Path("../../.env")
env_vars = [
    line.lstrip("export ").split("=")
    for line in envrc.read_text().splitlines()
    if line.startswith("export")
]
for k, v in env_vars:
    os.environ[k] = v
    print(f"Set {k}")
````

````python
username = os.getenv("API_USER")
client = PythonAnywhereClient(username)
````

````python
client.request(method="GET", url=client._base_url("cpu"))
````
